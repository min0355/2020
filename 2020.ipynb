### 1. __한국판 뉴딜 (New Deal)__ 이란 ? 

- 코로나 19 로 인해 최악의 경기 침체와 일자리 충격 등에 직면해 있는 상황에서 위기를 극복하고 코로나 이후 글로벌 경제를 선도하기 위해 
  마련된 국가 발전 전략임. 

- 20 년 4 월 22 일 제 5 차 비상 경제 회의에서 포스트 코로나 시대의 혁신 성장을 위한 대규모 국가 프로젝트로 뉴딜을 처음 언급하였으며,
  한국판 뉴딜 추진 전담 조직 (TF) 구성, 분야별 전문가 간담회, 민간제안 수렴 등을 거쳐 7 월 14 일, 제 7 차 비상경제 회의 겸 한국판 뉴딜
  국민 보고 대회를 통해 추진 계획이 발표됨. 
  
- 문재인 대통령은 국민 보고 대회 기조 연설에서 "한국판 뉴딜은 선도 국가로 도약하는 대한민국 대 전환 선언으로 추격형 경제에서 선도형 경제
  로, 탄소 의존에서 저 탄소 경제로, 불평등 사회에서 포용 사회로 대한민국을 근본적으로 바꾸어 새로운 100 년을 설계하는 것" 이라고 강조함.

### 2. 한국판 뉴딜을 바라보는 외부 시각

- 도널드 커크 (미국 기자) : 새로운 국면이다. 경제를 살리기 위한 많은 방안을 도입하는 것이며 중소 기업과 다양한 분야를 활성화하여 경기
  회복과 코로나 19 위기의 침체로부터 경제를 견인하기 위함이다. 여러 해 동안 많은 자금을 투입하여 그들을 격려하는 것이고 족벌 대 기업에 
  자금이 몰리는 것을 원하지 않음. 창의성과 혁신을 가진 중소기업을 발굴하고 격려하고 싶어함. 대기업을 무시하고 추진 할 수 있을지는 지켜
  봐야 함. 야심찬 계획이다.
  
  
- 앤드류 살먼 (영국 기자) : 멋진 브랜드 이긴 하지만 세계의 모든 정부가 시장에 돈을 쏟아 부으며 비슷한 일을 하고 있음. 실패하면 정치적
  부담이 있고 성공하면 큰 힘, 한국은 경제의 70 % 를 대외 무역에 의존하기 때문에 뉴딜 정책은 다른 나라의 정부 정책보다 중요함.
  왜냐하면 한국은 국내 정책에 따라 해외 시장에 대한 의존도가 달라지는 경제이기 때문이다.  


- 프랭크 스미스 (캐나다 기자) : 절박한 시기에는 과감한 조치가 필요하다고 판단, 코로나 바이러스는 사회 구조에 대한 심각한 도전이다.
  문재인 정부의 뉴딜 정책이 일자리와 사업을 잃은 사람들에게 더 강력한 사회 안전망을 제공하기 위해 개입하는 것은 타당함. 
  디지털 경제와 녹색 경제에 주력하는 것도 좋은 생각임. 전세계에서 일어나는 자동화라는 트랜드를 인지하고 이는 산업 혁명 이후 큰 패러다임
  제조업 분야에서는 더 이상 일자리가 늘어나고 있지 않기 때문에 고용 분야에서 보다 인본주의적 접근이 필요하다. 일하는 시간을 줄이고 
  일을 분담하는 유형의 조치를 취해야 할 것. 이번 뉴딜 정책은 경제 성장보다는 사회의 행복에 중점을 둔, 보다 장기적으로 지속 가능한 발전
  이라고 정의할 수 있다.
